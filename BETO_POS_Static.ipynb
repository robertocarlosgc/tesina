# Spanish BERT (BETO) + POS

```python
# https://huggingface.co/docs/transformers/installation
%pip install transformers
%pip install unidecode
```

```python
# https://huggingface.co/mrm8488/bert-spanish-cased-finetuned-pos?text=Mis+amigos+y+yo+estamos+pensando+en+viajar+a+Londres+este+verano.
from transformers import pipeline

# https://iq.opengenus.org/bert-cased-vs-bert-uncased/
# cased models keeps capital letters and accents
nlp_pos = pipeline(
    "ner",
    model="mrm8488/bert-spanish-cased-finetuned-pos",
    tokenizer=(
        'mrm8488/bert-spanish-cased-finetuned-pos',  
        {"use_fast": False}
))

text = 'Necesito comprar un kilogramo de naranjas, manzanas, tomate, limón, plátanos, cebollas y uvas para mañana'

pos_list = nlp_pos(text)
```

```python
df_pos = pd.DataFrame(pos_list)
df_pos
```

```python
query = df_pos.query("entity=='NC' and word != '[UNK]'")
query
```

```python
keywords = query["word"]
keywords
```

# Lemmatization

```python
%pip install -U pip setuptools wheel
%pip install -U spacy
!python -m spacy download es_core_news_sm
```

```python
# https://pharos.sh/python-para-pnl-tokenizacion-derivacion-y-lematizacion-con-la-biblioteca-spacy/
import spacy
sp = spacy.load('es_core_news_sm')
```

```python
keyword_list = []

for k in keywords:
  sentence = sp(k)
  for word in sentence:
    #print(word.text,  word.lemma_)
    keyword_list.append(word.lemma_)

keyword_list
```

```python
# https://www.geeksforgeeks.org/how-to-remove-string-accents-using-python-3/
import unidecode

df_final = pd.DataFrame()
LNB = 0

for l in keyword_list:
  normalizedKeyword = unidecode.unidecode(l)
  value = normalizedKeyword
  df_query = df_prov.query(f"Keyword==@value")
  
  if df_query.empty == False:
    df_final = df_final.append(df_prov.query(f"Keyword==@value"))
    LNB = LNB + 1
```

```python
df_final.sort_values(by="Score", ascending=False)
```

```python
original_tokenized_text = list(text.split(" "))

keyword_list_from_tokenized_text = []

for k in original_tokenized_text:
  sentence = sp(k)
  for word in sentence:
    #print(word.text,  word.lemma_)
    keyword_list_from_tokenized_text.append(word.lemma_)

keyword_list_from_tokenized_text
```

```python
LWT = 0

for s in keyword_list_from_tokenized_text:
  normalizedWord = unidecode.unidecode(s)
  value = normalizedWord
  df_tokenized_text = df_prov.query(f"Keyword==@value")

  if df_tokenized_text.empty == False:
    LWT = LWT + 1
```

#   Metrics
## Accuracy metrics
\begin{align}
        LNB = NumberOfLemmatized Nouns From BETO \\
        LWT = NumberOfLemmatized Words From Original Text \\
        \\
        percentage = \frac{LNB}{LWT}
        \\
        \\
        Where:
        \\
        LNB \leq LWT
    \end{align}

## Performance metrics
- Model size (MB): storage requirements
- Average latency (ms): average response time +- standard deviation

```python
# Metrics
print("LNB: ", LNB)
print("LNT: ", LWT)

percentage = LNB/LWT

print("percentage = LNB / LWT = ", percentage)
```
